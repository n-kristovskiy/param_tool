```python
TP = 0
TN = 0
FP = 0
FN = 0
with open('input.txt', 'r') as file:
    data_1 = list(map(int,file.readline().split()))
    data_2 = list(map(int,file.readline().split()))
for y_true, y_pred in zip(data_1, data_2):
    if y_true and y_pred:
        TP += 1
    if y_true == 0 and y_pred == 0:
        TN += 1
    if y_true == 0 and y_pred == 1:
        FP += 1
    if y_true == 1 and y_pred == 0:
        FN += 1
accuracy = (TP + TN)/(TP + TN + FP + FN)
print(f"{TP} {TN} {FP} {FN}\n{accuracy:.6f}")
```
