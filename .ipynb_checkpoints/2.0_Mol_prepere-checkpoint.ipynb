```python
trimer = 'moleculse/GRG_CIT.smiles'
monomer = 'moleculse/R_CIT.smiles'
trimer_smi = pt.open_smi(trimer)
monomer_smi = pt.open_smi(monomer)

print(f'your trimer: \n{trimer_smi}\n', 
      f'your monomer: \n{monomer_smi}', sep='')
```

```python
trimer_chem = Chem.MolFromSmiles(trimer_smi)
trimer_chem = Chem.AddHs(trimer_chem)

monomer_chem = Chem.MolFromSmiles(monomer_smi)
monomer_chem = Chem.AddHs(monomer_chem)
```

### Ограничения для рассчета зарядов

```python
symmetric_list = [[37,38],[34,35],[32,33],[30,31]]
charge_atom_dict = {8:-0.34790, 25:0.27470, 7:-0.26370, 24:0.15600, 5:0.73410, 6:-0.58940}
i_atoms_of_monomer = list(i_atoms_trimer)
charge_of_monomer = 0.0000
```

```python
def recalculate_charges( RESP_mol, symmetric_list: list, charge_dict: dict, loc_charge, loc_indices):
    constraints = psiresp.ChargeConstraintOptions()
    for pair_list in symmetric_list:
        constraints.add_charge_equivalence_constraint_for_molecule(RESP_mol,indices=pair_list)
    for index, charge in charge_dict.items():
        constraints.add_charge_sum_constraint_for_molecule(RESP_mol, charge=charge, indices = index)
    constraints.add_charge_sum_constraint_for_molecule(RESP_mol, charge=loc_charge, indices = loc_indices)
    job.charge_constraints = constraints
    charges_pll = job.run()
    return list(*charges_pll)
```

```python
charge_list = recalculate_charges(RESP_mol, symmetric_list, charge_atom_dict, charge_of_monomer, i_atoms_of_monomer)
```

```python
print('Заряд тримера:', sum(charge_list))
```

```python
ccc = charge_list
rounded_charges = list(map(lambda x: round(x, 4), ccc))
sum(rounded_charges)
```

```python
for i, atom in enumerate(RESP_mol.rdmol.GetAtoms()):

    atom.SetDoubleProp('PartialCharge', charge_list[i])
    atom.SetProp('atomNote',str("%4.4f" %  charge_list[i]))
    atom.SetAtomMapNum(atom.GetIdx())

display(RESP_mol.rdmol)
```

```python
sistem_ch = []
[sistem_ch.append(rounded_charges[index]) for index in  i_atoms_trimer ]
sum(sistem_ch)
```
